```python
import numpy as np

#Grupo 2

matriz_A = np.random.rand(20, 5)
print('Matriz A: ')
print(matriz_A)

matriz_B = np.random.normal(0, 10, size = (5, 7))
print('Matriz B: ')
print(matriz_B)

produto_matriz = matriz_A @ matriz_B
print('Poduto das matrizes: ')
print(produto_matriz)

print('3ª linha x 4ª coluna: ')
linha_x_coluna = np.dot(matriz_A[3:, ], matriz_B[:, 4])
print(linha_x_coluna)


matriz_final = np.dot(matriz_A, matriz_B[:, 4])
print('Matriz final: ')
print(matriz_final)
```
