```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

#Grupo 3

dados = pd.read_csv('Advertising.csv')

X = dados.iloc[:,1].values.reshape(-1,1)
Y = dados.iloc[:,2].values.reshape(-1,1)

lr = LinearRegression()
lr.fit(X,Y)
Ypredict = lr.predict(X)

declive = lr.coef_
intercesao = lr.intercept_

plt.plot (X, (X * declive + intercesao), color = 'g')
plt.title('Regressão Linear entre: \n')
plt.xlabel('\n Rádio')
plt.ylabel('Newspaper \n')
plt.scatter(X, Y)

declive[0]
d = declive

print(dados)
print(d)
```
